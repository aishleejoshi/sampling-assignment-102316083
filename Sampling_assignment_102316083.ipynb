```python
#a function to show before and after of various sampling techniques
def show_before_after(original_df, sampled_df, title):
    print(f"\n{'='*60}")
    print(title)
    print(f"{'='*60}")

    print("\nBefore Sampling:")
    print(original_df['Class'].value_counts())

    print("\nAfter Sampling:")
    print(sampled_df['Class'].value_counts())

```

```python
#simple random sampling
def simple_random_sampling(df, n):
    return df.sample(n=n, random_state=42)

```

```python
#systematic sampling
def systematic_sampling(df, k=5):
    return df.iloc[::k]

```

```python
#stratified sampling
def stratified_sampling(df, n):
    from sklearn.model_selection import StratifiedShuffleSplit

    split = StratifiedShuffleSplit(
        n_splits=1,
        train_size=n,
        random_state=42
    )

    for train_idx, _ in split.split(df, df['Class']):
        return df.iloc[train_idx]

```

```python
#cluster sampling
def cluster_sampling(df, n_clusters=5):
    df_copy = df.copy()
    df_copy['cluster'] = pd.cut(df_copy.index, n_clusters, labels=False)
    chosen_cluster = np.random.choice(df_copy['cluster'].unique())
    return df_copy[df_copy['cluster'] == chosen_cluster].drop('cluster', axis=1)

```

```python
#bootstrap sampling
def bootstrap_sampling(df, n):
    return df.sample(n=n, replace=True, random_state=42)

```

```python
sample_size = 1000

S1 = simple_random_sampling(balanced_df, sample_size)
S2 = systematic_sampling(balanced_df)
S3 = stratified_sampling(balanced_df, sample_size)
S4 = cluster_sampling(balanced_df)
S5 = bootstrap_sampling(balanced_df, sample_size)

```

```python
samples = {
    "Sampling 1: Simple Random Sampling": S1,
    "Sampling 2: Systematic Sampling": S2,
    "Sampling 3: Stratified Sampling": S3,
    "Sampling 4: Cluster Sampling": S4,
    "Sampling 5: Bootstrap Sampling": S5
}
```

```python
for name, sample_df in samples.items():
    show_before_after(balanced_df, sample_df, name)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
models = {
    "M1_Logistic": LogisticRegression(max_iter=1000),
    "M2_DecisionTree": DecisionTreeClassifier(random_state=42),
    "M3_RandomForest": RandomForestClassifier(random_state=42),
    "M4_KNN": KNeighborsClassifier(),
    "M5_SVM": SVC()
}
```

```python
samples = {
    "Sampling1": S1,
    "Sampling2": S2,
    "Sampling3": S3,
    "Sampling4": S4,
    "Sampling5": S5
}
```

```python
results = pd.DataFrame(
    index=models.keys(),
    columns=samples.keys()
)
```

```python
for samp_name, samp_df in samples.items():

    X_s = samp_df.drop("Class", axis=1)
    y_s = samp_df["Class"]

    for model_name, model in models.items():
        scores = cross_val_score(
            model,
            X_s,
            y_s,
            cv=5,
            scoring="accuracy"
        )

        results.loc[model_name, samp_name] = round(scores.mean() * 100, 2)
```

```python
print("\nFinal Accuracy Table:")
print(results)
```

```python
results_numeric = results.astype(float)
```

```python
best_sampling_per_model = pd.DataFrame({
    "Best_Sampling_Technique": results_numeric.idxmax(axis=1),
    "Best_Accuracy (%)": results_numeric.max(axis=1)
})
```

```python
print("\nBest Sampling Technique for Each Model:")
print(best_sampling_per_model)
```
